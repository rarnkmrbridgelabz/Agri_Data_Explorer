# 3.Data Cleaning

## 3.1 Renaming column names

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.info()
```

```python
df.head(5)
```

## 3.2 Standardize units to ensure consistency

```python
#                        Given in                                  Convert to                      How
# --------------------------------------------------------------------------------------------------------------------
# 1. Area      :         1000 hectares                ->            hectares                Multiplying by 1000
# 2. Production:         1000 tons                    ->            tons                    Multiplying by 1000
# 3. Yield     :         kg per hectare               ->            No conversion needed            -
```

```python
area_column = [col for col in df.columns if 'area_(1000_ha)' in col]
production_column = [col for col in df.columns if 'production_(1000_tons)' in col]
yield_column = [col for col in df.columns if 'yield_(kg_per_ha)' in col]
```

```python
#convert area column
for col in area_column:
    df[col] *=1000
```

```python
# convert production column
for col in production_column:
    df[col] *=1000
```

```python
# Renaming Columns in the DataFrame using Dictionary Comprehension
rename_area = {col : col.replace('(1000_ha)','hectares') for col in area_column}
df.rename(columns=rename_area,inplace=True)

rename_production = {col : col.replace('(1000_tons)','tons') for col in production_column}
df.rename(columns=rename_production,inplace=True)

rename_yield= {col : col.replace('(kg_per_ha)','kg/hectare') for col in yield_column}
df.rename(columns=rename_yield,inplace=True)
```

## 3.4 Addressing missing values

```python
# S.No          FOR                 USE
# ----------------------------------------------
#  1     Numerical columns          mean()
#  2     Categorical columns        mode()
```

### 3.4.1 Create a function to handle missing values

```python
def handling_missing_values(df):
    for col in df.columns:
        if col.dtype == "object": #----------------------------->Categorical columns
            df.fillna(df[col].mode()[0], inplace=True)
        else:#-------------------------------------------------->Numerical columns
            df.fillna(df[col].mean(), inplace=True)
    return df   
```
